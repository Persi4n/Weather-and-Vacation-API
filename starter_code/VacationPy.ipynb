## Starter Code to Import Libraries and Load the Weather and Coordinates Data

```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests

# Import API key
from api_keys import geoapify_key
```

```python
# Load the CSV file created in Part 1 into a Pandas DataFrame
city_data_df = pd.read_csv("../output_data/cities.csv")

# Display sample data
city_data_df.head()
```

---

### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
%%capture --no-display

# Creating a map
map_plot = city_data_df.hvplot.scatter(
    x='Lng',
    y='Lat',
    c='Humidity',
    size='Humidity',
    hover_cols=['City', 'Country', 'Humidity'],
    title="City Map with Humidity",
    colorbar=True,
    cmap='viridis',
    width=800,
    height=500
)

map_plot
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
#Ideal weather conditions.
ideal_conditions_df = city_data_df[
    (city_data_df['Max Temp'] < 27) & 
    (city_data_df['Max Temp'] > 21) & 
    (city_data_df['Wind Speed'] < 4.5) & 
    (city_data_df['Cloudiness'] == 0)
]

ideal_conditions_df
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Hotel Data Frame
hotel_df = ideal_conditions_df.copy()
hotel_df['Hotel Name'] = ""
hotel_df
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
radius = 10000  
params = {
    "apiKey": geoapify_key,
    "filter": "category=hotel",
    "bias": "radius",
    "limit": 1
}

print("Starting hotel search")

for index, row in hotel_df.iterrows():
    lat = row['Lat']
    lng = row['Lng']
    params["bias"] = f"radius:{lat},{lng},{radius}"
    
    base_url = "https://api.geoapify.com/v2/places"
    response = requests.get(base_url, params=params)
    data = response.json()
    
    try:
        hotel_name = data["features"][0]["properties"]["name"]
        hotel_df.at[index, "Hotel Name"] = hotel_name
    except (KeyError, IndexError):
        hotel_df.at[index, "Hotel Name"] = "No hotel found"
    
    print(f"{hotel_df.at[index, 'City']} - nearest hotel: {hotel_df.at[index, 'Hotel Name']}")
```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
%%capture --no-display

map_plot_with_hotels = hotel_df.hvplot.scatter(
    x='Lng',
    y='Lat',
    c='Humidity',
    size='Humidity',
    hover_cols=['City', 'Country', 'Humidity', 'Hotel Name'],
    title="City Map with Humidity and Hotels",
    colorbar=True,
    cmap='viridis',
    width=800,
    height=500
)

map_plot_with_hotels
```
